```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon, norm
import statistics
from math import sqrt
```

# Экспоненциальное распределение
Дисперсия экспоненциального распределения $D = \lambda^{-2}$. 
Так как я использую стандартное экспоненциальное распределение, то параметр $\lambda$ равен 1

```python
#Построение гистограмы
r = expon.rvs(size=1000)
plt.hist(r, density=True)
#Построение графика плотности распределения
expon_rv = expon()
x = np.linspace(0, 5, 1000)
plt.plot(x, expon.pdf(x), label='theoretical PDF')
plt.legend(loc='best', frameon=False)
plt.ylabel('fraction of samples')
plt.xlabel('x')
plt.show()
```

```python
#Создание по 1000 выборок для каждого размера (5, 10, 50, 100)
size_five_array = []
size_ten_array = []
size_fifty_array = []
size_hundred_array = []
for j in range(1000):
    size_five_array.append(expon.rvs(size=5))
    size_ten_array.append(expon.rvs(size=10))
    size_fifty_array.append(expon.rvs(size=50))
    size_hundred_array.append(expon.rvs(size=100))
```

```python
#Массивы со средними выборок
size_five_sample_mean = [statistics.mean(size_five_array[i]) for i in range (1000)]
size_ten_sample_mean = [statistics.mean(size_ten_array[i]) for i in range (1000)]
size_fifty_sample_mean = [statistics.mean(size_fifty_array[i]) for i in range (1000)]
size_hundred_sample_mean = [statistics.mean(size_hundred_array[i]) for i in range (1000)]
```

# Гистограмма и нормальное распределение для средних выборки размера 5

```python
plt.hist(size_five_sample_mean, density=True)
plt.ylabel('fraction of samples')
plt.xlabel('x')
#Среднее
m = statistics.mean(r)
#Нормальное распределение при n = 5
n5 = norm(m, 1/ sqrt(5))
#Построение графика плотности нормального распределения
plt.plot(x, n5.pdf(x))
```

# Гистограмма и нормальное распределение для средних выборки размера 10

```python
plt.hist(size_ten_sample_mean, density=True)
plt.ylabel('fraction of samples')
plt.xlabel('x')
#Нормальное распределение при n = 10
n10 = norm(m, 1/sqrt(10))
#Построение графика плотности нормального распределения
plt.plot(x, n10.pdf(x))
```

# Гистограмма и нормальное распределение для средних выборки размера 50

```python
plt.hist(size_fifty_sample_mean, density=True)
plt.ylabel('fraction of samples')
plt.xlabel('x')
#Нормальное распределение при n = 50
n50 = norm(m, 1/sqrt(50))
#Построение графика плотности нормального распределения
plt.plot(x, n50.pdf(x))
```

# Гистограмма и нормальное распределение для средних выборки размера 100

```python
plt.hist(size_hundred_sample_mean, density=True)
plt.ylabel('fraction of samples')
plt.xlabel('x')
#Нормальное распределение при n = 100
n100 = norm(m, 1/sqrt(100))
#Построение графика плотности нормального распределения
plt.plot(x, n100.pdf(x))
```

При увеличении n увеличивается точность аппроксимации, график распределения начинает быть похожим на нормальное распределение. Из-за сильной скошенности экспоненциального распределения, это достигается при больших n, чем требуется в случае симметричного распределения
