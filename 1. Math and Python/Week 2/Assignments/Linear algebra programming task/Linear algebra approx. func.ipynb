# Приближение многочленом первой степени

```python
import numpy as np
from scipy.linalg import solve
import math
import matplotlib.pyplot as plt
```

```python
b1 = math.sin(1 / 5) * math.exp(1 / 10) + 5 * math.exp(-1 / 2)
b2 = math.sin(15 / 5) * math.exp(15 / 10) + 5 * math.exp(-15 / 2)
```

```python
def f(x):
    return math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp(-x / 2)
```

```python
a = np.array([[1, 1], [1, 15]])
b = np.array([b1, b2])
x = solve(a, b)
def mn1(x):
    return 3.43914511 - 0.18692825 * x
exp = []
xd = []
mn = []
for i in np.arange(0, 15, 0.01):
    exp.append(f(i))
    mn.append(mn1(i))
    xd.append(i)
plt.plot(xd, exp)
plt.plot([1, 15], [f(1), f(15)])
```

# Приближение многочленом второй степени

```python
a = np.array([[1, 1, 1], [1, 8, 64], [1, 15, 225]])
b = np.array([f(1), f(8), f(15)])
x = solve(a, b)
def mn2(x):
    return 3.32512949 - 0.06531159 * x - 0.00760104 * x * x
mn = []
exp = []
xd = []
for i in np.arange(0, 15, 0.01):
    exp.append(f(i))
    mn.append(mn2(i))
    xd.append(i)
plt.plot(xd, exp)
plt.plot(xd, mn)
```

# Приближение многочленом третьей степени
